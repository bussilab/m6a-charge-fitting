```python
import plumed
import numpy as np

Bias=[]
for k in range(0,16):
    bias=plumed.read_as_pandas("bias"+str(k)+"_stride10")
    print(k)
    Bias.append(bias["metad.bias"])
Bias=np.array(Bias)
Bias=np.transpose(Bias)
```

```python
import numpy as np
import panedr
from bussilab import wham

em = "lam12/ener_trj_conc.edr"
df = panedr.edr_to_df(em)
df[u'Potential']
#Ene2d=np.ones(len(df)*16).reshape((len(df),16))




Ene2d=np.zeros(np.shape(Bias))

for i in range(16):
   em = "lam"+str(i)+"/ener_trj_conc.edr"
   df = panedr.edr_to_df(em)
   arr=np.array(df[u'Potential'])
   print(i,np.shape(arr))
   for j in range((len(df))):
        Ene2d[j][i]+=Bias[j][i]+arr[j]
```

```python
bias=Ene2d-Ene2d[:,0][:,np.newaxis]
a =wham.wham(bias=bias, T=2.494, maxiter=10000)
2.494*(a[u'logZ'][15]-a[u'logZ'][0])
```

```python
np.save("logW",a.logW)
np.save("logZ",a.logZ)
np.save("DeE",Ene2d[:,15]-Ene2d[:,0])
np.save("Bias",Bias)
```

```python
Nene=200
DeGs=[]

#Reading Demux indexes, they are needed to correctly count how much times the extracted trajecotries spend in each Hamiltonian system
replica_temp=np.loadtxt("replica_temp.xvg")
replica_temp=replica_temp[:,1:]
replica_temp=replica_temp.astype(int)


for i in range(Nene):
    bias=Ene2d-Ene2d[:,0][:,np.newaxis]
    bias=bias[:160000]
    
    tr_w=np.zeros(16)
    sample=np.random.choice(16,size=16)
    for k in range(len(sample)):
        L=np.bincount(replica_temp[:,sample[k]],minlength=16)
        for j in range(len(L)):
            tr_w[j]+=L[j]
            
    bias=bias.reshape((16,-1,16))[sample,:,:].reshape((-1,16))
    a =wham.wham(bias=bias, T=2.494, traj_weight=tr_w, maxiter=10000, logZ=a.logZ)
    DeGs.append(2.494*(a[u'logZ'][15]-a[u'logZ'][0]))
    
np.std(DeGs)
```
